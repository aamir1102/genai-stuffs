```python
from pptx import Presentation
from langchain_core.documents import Document

def load_pptx_from_python_pptx(file_path):
    """Loads a PPTX file using python-pptx and converts it to a list of LangChain Documents."""
    prs = Presentation(file_path)
    langchain_documents = []
    
    for i, slide in enumerate(prs.slides):
        slide_text = []
        slide_title = f"Slide {i + 1}"  # Default title
        
        # 1. Safely check for a title placeholder (Placeholder idx 0 is usually the title)
        title_shape = None
        for shape in slide.placeholders:
            if shape.placeholder_format.idx == 0:
                title_shape = shape
                break

        if title_shape and hasattr(title_shape, 'text') and title_shape.text:
            slide_title = title_shape.text
            
        # 2. Extract text from all shapes
        for shape in slide.shapes:
            if hasattr(shape, "text_frame") and shape.text_frame:
                text = shape.text_frame.text
                if text:
                    slide_text.append(text)
        
        # 3. Create a Document for each slide
        doc = Document(
            page_content="\n".join(slide_text),
            metadata={
                "source": file_path,
                "slide_number": i + 1, 
                "title": slide_title
            }
        )
        langchain_documents.append(doc)

    return langchain_documents

# Run the code using the custom function
# Make sure your file 'GPT Experiment Summary (1).pptx' is in the same directory as this script.
ppt_documents = load_pptx_from_python_pptx("GPT Experiment Summary (1).pptx")
print(ppt_documents)
```

```python
import os
# NEW: Import text splitter from the dedicated package
from langchain_text_splitters import RecursiveCharacterTextSplitter 
# NEW: Loaders are typically in the community package
from langchain_community.document_loaders import PyPDFLoader 
# NEW: LLMs (like OpenAI) are in the dedicated integration package

# These core chains and prompts are still in the main 'langchain' package
from langchain.chains.summarize import load_summarize_chain
from langchain.prompts import PromptTemplate

# --- CONFIGURATION ---

# 1. Set your OpenAI API Key
# Replace 'YOUR_API_KEY' or ensure it's set as an environment variable (os.environ["OPENAI_API_KEY"])
# os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY" 

# 2. Path to your PDF file
pdf_path = "rag-test.pdf" 
# NOTE: Make sure to have a PDF named 'your_document.pdf' in the same directory, or change the path.

# 3. Text Splitter Settings
CHUNK_SIZE = 1500
CHUNK_OVERLAP = 150

# --- STEP 1: LOAD AND SPLIT THE DOCUMENT ---

try:
    print(f"Loading document: {pdf_path}")
    loader = PyPDFLoader(pdf_path)
    pages = loader.load()
    
    # Split the document into chunks
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHUNK_SIZE,
        chunk_overlap=CHUNK_OVERLAP, 
    )

    docs = text_splitter.split_documents(pages)
    print(f"Original pages: {len(pages)}")
    print(f"Number of chunks created: {len(docs)}")

except FileNotFoundError:
    print(f"Error: The file '{pdf_path}' was not found. Please check the path.")
    exit()


# --- STEP 2: DEFINE THE SUMMARIZATION CHAIN ---

# 1. Initialize the LLM (using a larger context model for better results)
# You can change 'gpt-3.5-turbo-16k' to another model (e.g., 'gpt-4')

# 2. Define Prompts for the Map-Reduce Process

# Map Prompt (Summarize each chunk)
map_prompt = """
You are an expert summarizer. Summarize the following piece of text from a document.
Do not add any external information. Keep the summary concise and focused on the key points of this section.
TEXT: "{text}"
SUMMARY:
"""
MAP_PROMPT = PromptTemplate(template=map_prompt, input_variables=["text"])

# Reduce Prompt (Consolidate all summaries)
reduce_prompt = """
You are an expert at merging summaries. Given the following set of individual summaries,
create a single, coherent, and comprehensive final summary of the entire document.
FINAL SUMMARIES: "{text}"
FINAL SUMMARY:
"""
REDUCE_PROMPT = PromptTemplate(template=reduce_prompt, input_variables=["text"])


# 3. Create the Summarization Chain
# 'map_reduce' is essential for documents that are too long for a single LLM call.
summarize_chain = load_summarize_chain(
    llm,
    chain_type="map_reduce",
    map_prompt=MAP_PROMPT,
    combine_prompt=REDUCE_PROMPT,
    verbose=True # Shows the steps being executed
)

# --- STEP 3: RUN THE CHAIN ---

print("\n--- Starting Map-Reduce Summarization ---")
final_summary = summarize_chain.run(docs)

print("\n--- Final PDF Summary ---")
print(final_summary)
```
